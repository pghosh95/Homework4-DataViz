Data Analysis of PyCity Schools

1) Students from "charter" schools are more likely to outperform students from "district schools.

2) Students from schools with a lower spending range per student typicially outperform students from schools with a higher spending range per student.

3) Students from all schools typically have a higher higher reading score than their math score


```python
import pandas as pd
import numpy as np
```

```python
school_data = "Documents/schools_complete.csv"
student_data = "Documents/students_complete.csv"
student_df=pd.read_csv(student_data)
school_df=pd.read_csv(school_data)
```

```python
combined_df = pd.merge(student_df, school_df, how = "left", on = "school_name")
combined_df.head()
```

```python
school_df.columns
```

# District Analysis

```python
Total_Schools = len(school_df["School ID"])
Total_Students = school_df["size"].sum()
Total_Budget= school_df["budget"].sum()
Average_Math_Score = student_df["math_score"].mean()
Average_Reading_Score = student_df["reading_score"].mean()

Total_Stud=len(student_df["Student ID"])
Pass_Math=(len(student_df[student_df["math_score"] >= 70])/ Total_Stud)
Pass_Reading=(len(student_df[student_df["reading_score"] >= 70])/ Total_Stud)

overall_score=((Pass_Math+Pass_Reading)/2)*100
```

```python
Passing_Math_p = (len(student_df[student_df["math_score"] >= 70])/ Total_Stud)*100
Passing_Reading_p = (len(student_df[student_df["reading_score"] >= 70])/ Total_Stud)*100
Overall_Passing_Rate = overall_score
```

```python
District_Summary=pd.DataFrame({"Total Schools": [Total_Schools],
                              "Total Students": [Total_Stud],
                              "Total Budget": [Total_Budget],
                              "Average Math Score": [Average_Math_Score],
                              "Average Reading Score": [Average_Reading_Score],
                              "% Passing Math": [Passing_Math_p],
                              "% Passing Reading": [Passing_Reading_p],
                              "Overall Passing Rate": [Overall_Passing_Rate]})
District_Summary["Total Students"] = District_Summary["Total Students"].map("{: ,}".format)
District_Summary["Total Budget"] = District_Summary["Total Budget"].map("${:,.2f}".format)

District_Summary
```

```python
school_df
```

# School Summary

```python
types = school_df.set_index(["school_name"])["type"]

school_counts = combined_df["school_name"].value_counts()
school_budget = combined_df.groupby(["school_name"]).mean()["budget"]
school_capita = school_budget/school_counts

avg_math = combined_df.groupby(["school_name"]).mean()["math_score"]
avg_reading = combined_df.groupby(["school_name"]).mean()["reading_score"]

```

```python
pass_math2 = combined_df[(combined_df["math_score"] >= 70)]
pass_read2 = combined_df[(combined_df["reading_score"] >= 70)]
pass_math_p = pass_math2.groupby(["school_name"]).count()["student_name"]/school_counts *100
pass_reading_p = pass_read2.groupby(["school_name"]).count()["student_name"]/school_counts *100
overall = (pass_math_p + pass_reading_p)/2
```

```python

school_summary = pd.DataFrame({
    "School Type": types,
    "Total Students": school_counts,
    "Per Student Budget": school_capita,
    "Total School Budget": school_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_reading,
    "% Passing Math": pass_math_p,
    "% Passing Reading": pass_reading_p,
    "Overall Passing Rate": overall
})
school_summary
```

```python
school_summary = school_summary[['School Type', 
                          'Total Students', 
                          'Total School Budget', 
                          'Per Student Budget', 
                          'Average Math Score', 
                          'Average Reading Score',
                          '% Passing Math',
                          '% Passing Reading',
                          'Overall Passing Rate']]
```

```python
school_summary.style.format({'Total Students': '{:,}', 
                          "Total School Budget": "${:,}", 
                          "Per Student Budget": "${:.0f}",
                          'Average Math Score': "{:.1f}", 
                          'Average Reading Score': "{:.1f}", 
                          })
```

# Top Performing Schools

```python
TOP5=school_summary.sort_values("Overall Passing Rate", ascending = False)


TOP5.style.format({'Total Students': '{:,}',
                           "Total School Budget": "${:,}", 
                           "Per Student Budget": "${:.0f}", 
                           "% Passing Math": "{:.1%}", 
                           "% Passing Reading": "{:.1%}", 
                           "Overall Passing Rate": "{:.1%}"})



TOP5.head()
```

# Bottom Performing Schools

```python
Bottom_5 = TOP5.tail()

Bottom_5 = Bottom_5.sort_values('Overall Passing Rate')
Bottom_5.style.format({'Total Students': '{:,}', 
                       "Total School Budget": "${:,}", 
                       "Per Student Budget": "${:.0f}", 
                       "% Passing Math": "{:.1%}", 
                       "% Passing Reading": "{:.1%}", 
                       "Overall Passing Rate": "{:.1%}"})
Bottom_5.head()
```

# Math Scores by Grade

# Scores by School Spending

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

school_summary["Spending Ranges (Per Student)"] = pd.cut(school_capita, spending_bins, labels=group_names)

spending_math_scores = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_rate = (spending_math_scores + spending_reading_scores) / 2
```

```python
spending_summary = pd.DataFrame({"Average Math Score" : spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Math": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate})

spending_summary = spending_summary[["Average Math Score",
                                     "Average Reading Score",
                                     "% Passing Math", "% Passing Reading",
                                     "% Overall Passing Rate"]]
spending_summary
```

# Scores by School Size

```python
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_summary["School Size"] = pd.cut(school_summary["Total Students"], size_bins, labels=group_names)
```

```python
s_size_math_scores = school_summary.groupby(["School Size"]).mean()["Average Math Score"]
s_size_reading_scores = school_summary.groupby(["School Size"]).mean()["Average Reading Score"]
s_size_passing_math = school_summary.groupby(["School Size"]).mean()["% Passing Math"]
s_size_passing_reading = school_summary.groupby(["School Size"]).mean()["% Passing Reading"]
s_size_overall_passing_rate = (s_size_passing_math + s_size_passing_reading) / 2
```

```python
school_size_summary = pd.DataFrame({"Average Math Score" : s_size_math_scores,
                             "Average Reading Score": s_size_reading_scores,
                             "% Passing Math": s_size_passing_math,
                             "% Passing Reading": s_size_passing_reading,
                             "% Overall Passing Rate": s_size_overall_passing_rate})


school_size_summary = school_size_summary[["Average Math Score",
                             "Average Reading Score",
                             "% Passing Math", "% Passing Reading",
                             "% Overall Passing Rate"]]
school_size_summary
```
